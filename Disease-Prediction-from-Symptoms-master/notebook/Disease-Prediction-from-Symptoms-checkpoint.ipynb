```python
# Import Dependencies
import csv
import pandas as pd
import numpy as np
from collections import defaultdict
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Process Disease and Symptom Names
def process_data(data):
    data_list = []
    data_name = data.replace('^','_').split('_')
    n = 1
    for names in data_name:
        if (n % 2 == 0):
            data_list.append(names)
        n += 1
    return data_list
```

```python
# Data Cleanup
import pandas as pd
from collections import defaultdict

# Define a function to process data (you can replace this with your actual data processing logic)
def process_data(data):
    # Replace this with your data processing logic
    return data

# Assuming you have loaded your data into the 'data' DataFrame

# Initialize data structures
disease_list = []
disease_symptom_dict = defaultdict(list)
disease_symptom_count = defaultdict(int)

# Iterate through the DataFrame rows
for idx, row in data.iterrows():
    
    # Get the Disease Names (using 'Unnamed: 0' column)
    if (row['Unnamed: 0'] != "\xc2\xa0") and (row['Unnamed: 0'] != ""):
        disease = row['Unnamed: 0']
        disease_list = process_data(data=disease)
        count = row['Count of Disease Occurrence']

    # Get the Symptoms Corresponding to Diseases (using 'Symptom' column)
    if (row['Symptom'] != "\xc2\xa0") and (row['Symptom'] != ""):
        symptom = row['Symptom']
        symptom_list = process_data(data=symptom)
        
        # Update disease_symptom_dict
        for d in disease_list:
            disease_symptom_dict[d].extend(symptom_list)
            
        # Update disease_symptom_count
        for d in disease_list:
            disease_symptom_count[d] += count

# Now, 'disease_symptom_dict' contains diseases and their corresponding symptoms,
# and 'disease_symptom_count' contains the count of occurrences for each disease.


```

```python
# Save cleaned data as CSV
f = open('./dataset/cleaned_data.csv', 'w')

with f:
    writer = csv.writer(f)
    for key, val in disease_symptom_dict.items():
        for i in range(len(val)):
            writer.writerow([key, val[i], disease_symptom_count[key]])
```

```python
# Read Cleaned Data as DF
import pandas as pd

# Specify the encoding when reading the CSV file
df = pd.read_csv('./dataset/cleaned_data.csv', encoding='ISO-8859-1')

# Rename columns
df.columns = ['disease', 'symptom', 'occurrence_count']

# Display the first few rows of the DataFrame
df.head()
```

```python
# Remove any rows with empty values
df.replace(float('nan'), np.nan, inplace=True)
df.dropna(inplace=True)
```

```python
# Encode the Labels
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(df['symptom'])
print(integer_encoded)
```

```python
# One Hot Encode the Labels
onehot_encoder = OneHotEncoder(sparse=False)
integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
print(onehot_encoded)
```

```python
onehot_encoded[0]
```

```python
len(onehot_encoded[0])
```

```python
cols = np.asarray(df['symptom'].unique())
cols
```

```python
# Create a new dataframe to save OHE labels
df_ohe = pd.DataFrame(columns = cols)
df_ohe.head()
```

```python
for i in range(len(onehot_encoded)):
    df_ohe.loc[i] = onehot_encoded[i]
```

```python
df_ohe.head()
```

```python
len(df_ohe)
```

```python
# Disease Dataframe
df_disease = df['disease']
df_disease.head()
```

```python
# Concatenate OHE Labels with the Disease Column
df_concat = pd.concat([df_disease,df_ohe], axis=1)
df_concat.head()
```

```python
df_concat.drop_duplicates(keep='first',inplace=True)
```

```python
df_concat.head()
```

```python
len(df_concat)
```

```python
cols = df_concat.columns
cols
```

```python
cols = cols[1:]
```

```python
# Since, every disease has multiple symptoms, combine all symptoms per disease per row
df_concat = df_concat.groupby('disease').sum()
df_concat = df_concat.reset_index()
df_concat[:5]
```

```python
len(df_concat)
```

```python
df_concat.to_csv("./dataset/training_dataset.csv", index=False)
```

```python
# One Hot Encoded Features
X = df_concat[cols]

# Labels
y = df_concat['disease']
```
